```python
import numpy
import matplotlib.pyplot as plt
```

```python
# Pattern (between zero and 255)
number = 254
```

```python
# convert decimal number into binary number
output_pattern = [int(x) for x in numpy.binary_repr(number, width=8)]
output_pattern
```

```python
input_pattern = numpy.zeros([8,3])
input_pattern
```

```python
A = range(8)
A[0]
```

```python
for i in range(8):
    input_pattern[i,:]= [int(x) for x in numpy.binary_repr(7-i, width=3)]
print(input_pattern)
```

```python
# set up columns and rows for the 1c
#columns should be an odd number
columns = 201
rows = int(columns/2)+1
rows
```

```python
canvas = numpy.zeros([rows,columns+2])
##--Option 1
##convert the middle element of the top row to 1
canvas[0,int(columns/2+1)]=1
##--Option 2
##convert the top row into a random binary array
#canvas[0] = numpy.random.randint(2, size=columns+2)
##Show result
canvas
```

```python
for i in numpy.arange(0,rows-1):
    for j in numpy.arange(0, columns):
        for k in range(8):
            if numpy.array_equal(input_pattern[k,:], canvas[i,j:j+3]):
                canvas[i+1,j+1] = output_pattern[k]
```

```python
plt.figure(figsize = (300,20))
plt.imshow(canvas[:,1:columns+1], cmap='Greys', interpolation='nearest')
plt.title("Elementary 1D Cellular Automata Rule {}".format(number))
plt.show()
```
